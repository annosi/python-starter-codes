```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/parsed.csv')
df.sample(5)
```

**Find the 95th percentile of earthquake magnitude in Japan using the mb
magnitude type.**

```python
df[(df.parsed_place=='Japan')&(df.magType=='mb')].describe(percentiles=[0.95]).at['95%','mag']
```

```python
#above method is too long.
#use the python code for percentiles which is quantile
df[(df.parsed_place=='Japan')&(df.magType=='mb')]['mag'].quantile(0.95)
```

**Find the percentage of earthquakes in Indonesia that were coupled with tsunamis.**

```python
all_earthquake = df[df.parsed_place=='Indonesia']['type'].count()
print(all_earthquake)
with_tsunami = df[(df.parsed_place=='Indonesia') & (df.tsunami==1)]['type'].count()
print(with_tsunami)
print(100*with_tsunami/all_earthquake,'%')
```

**Add a column indicating whether the earthquake happened in a country or US
state that is on the Ring of Fire. Use Alaska, Antarctica (look for Antarctic), Bolivia,
California, Canada, Chile, Costa Rica, Ecuador, Fiji, Guatemala, Indonesia, Japan,
Kermadec Islands, Mexico (be careful not to select New Mexico), New Zealand,
Peru, Philippines, Russia, Taiwan, Tonga, and Washington.**

```python
df['Ring of Fire'] = (
    (df.parsed_place=='Alaska')|(df.parsed_place=='Antarctic')|(df.parsed_place=='Bolivia')|(df.parsed_place=='California')|
    (df.parsed_place=='Canada')|(df.parsed_place=='Chile')|(df.parsed_place=='Costa Rica')|(df.parsed_place=='Ecuador')|
    (df.parsed_place=='Fiji')|(df.parsed_place=='Guatemala')|(df.parsed_place=='Indonesia')|(df.parsed_place=='Japan')|
    (df.parsed_place=='Kermadec Islands')|(df.parsed_place=='Mexico')|(df.parsed_place=='New Zealand')|
    (df.parsed_place=='Peru')|(df.parsed_place=='Philippines')|(df.parsed_place=='Russia')|(df.parsed_place=='Taiwan')|
    (df.parsed_place=='Tonga')|(df.parsed_place=='Washington')
)
df.head(10)
```

**Calculate the number of earthquakes in the Ring of Fire locations and the number
outside of them.**

```python
in_location = df[df['Ring of Fire']==True]['type'].count()
print(in_location)
out_location = df[df['Ring of Fire']==False]['type'].count()
print(out_location)
#do the below simply since you know it contains only true and false
test = df['Ring of Fire'].value_counts()
print(test)
```

**Find the tsunami count along the Ring of Fire.**

```python
tsunami_count = df[df['Ring of Fire']==True]['tsunami'].sum()
print(tsunami_count)
```

```python
#OR
tsunami_count2 = df[(df['Ring of Fire']==True) & (df.tsunami==1)]['tsunami'].count()
print(tsunami_count2)
```

```python
#best solution simple and short
df.loc[df['Ring of Fire'], 'tsunami'].sum()
```
